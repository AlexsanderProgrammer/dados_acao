COTAÇÃO AÇÕES

```python
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
```

PEGANDO OS DADOS DE UMA EMPRESA

```python
cotacao_bovespa = web.DataReader('^BVSP', data_source= 'yahoo', start = "01-01-2021", end = "01-01-2022" )
cotacao_bovespa["Adj Close"].plot()
plt.show()
display(cotacao_bovespa)
```

PEGANDO DADOS DE VARIAS EMPRESAS

```python
tabela_empresas = pd.read_excel("Empresas.xlsx")#LENDO O ARQUIVO DE EXCEL COM O PANDAS
display(tabela_empresas)


for empresas in tabela_empresas['Empresas']:#PARA EMPRESA NA TABELA_EMPRESA PEGA O INDICE OU COLUNA[EMPRESAS]
    print(empresas)
    cotacao = web.DataReader(f'{empresas}.SA', data_source= 'yahoo', start = "01-01-2021", end = "01-01-2022" )#LENDO OS DADOS DA AÇÃO NO SITE COM O DATAREADER
    cotacao["Adj Close"].plot( figsize = (15,10))#CRIANDO UM GRAFICO COM PRAMETRO DE TAMANHO LARGURA/ALTURA RESPECTIVAMENTE
    plt.show()#MOSTRANDO O GRAFICO
    display(cotacao)
```
